```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME *"2020"*

```python
#Cargar archivo
data1=pd.read_csv("Gastos y costos 2020.csv")
```

```python
data1.info()
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
data2020= data1.drop(['Unnamed: 0'], axis=1)
data2020
```

```python
data2020.info()
```

```python
cualitativas2020=data2020.iloc[ : , [0,1,2,3,4,5,6,7,15,16,17]]
cualitativas2020
```

```python
cuantitativas2020=data2020.iloc[ : , [8,9,10,11,12,13,14,18]]
cuantitativas2020
```

```python
#Para identificar los outlayers

#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize=(15, 8))
data2020.plot(kind='box', vert=False)
plt.title("Valores Atípicos del DataFrame 2020")
plt.show()   # dibujamos el diagrama
```

```python
#Método aplicando desviación estándard. Encuentro los valores extremos
y=cuantitativas2020
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print("Límite Superior permitido", Limite_Superior)
print("Límite Inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los autliers se convierten en nulos en el DataFrame
data2020_1 = cuantitativas2020[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2020_1
```

```python
#Cooroboramos valores nulos del DataFrame
valores_nulos2020=data2020_1.isnull().sum()
valores_nulos2020
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean2020=data2020_1.copy()
data_clean2020=data_clean2020.fillna(round(data2020_1.mean(),1))
data_clean2020
```

```python
#Cooroboramos valores nulos del DataFrame LIMPIO
valores_nulos2020=data_clean2020.isnull().sum()
valores_nulos2020
```

```python
#Unimos la columnas CUALITATIVAS con el dataframe

Datos_limpios2020 = pd.concat([cualitativas2020, data_clean2020], axis=1)
Datos_limpios2020
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos2020=Datos_limpios2020.isnull().sum()
valores_nulos2020
```

```python
#Convertir DataFrame a CSV
Datos_limpios2020.to_csv("Valores Atipicos 2020 LIMPIOS.csv")
```

PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME *2021*

```python
#Cargar archivo
data2=pd.read_csv("Gastos y costos 2021.csv")
```

```python
data2.info()
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
data2021= data2.drop(['Unnamed: 0'], axis=1)
data2021
```

```python
data2021.info()
```

```python
#Cualitativas : Object
cualitativas2021=data2021.iloc[ : , [0,1,2,3,4,5,6,7,15,16,17]]
cualitativas2021
```

```python
#Cuantitativas : Float
cuantitativas2021=data2021.iloc[ : , 8:14]
cuantitativas2021
```

```python
#Para identificar los outlayers

#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize=(15, 8))
data2021.plot(kind='box', vert=False)
plt.title("Valores Atípicos del DataFrame 2021")
plt.show()   # dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas2021
percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr=percentile75-percentile25

Limite_Superior_iqr2021 = percentile75 + 1.5*iqr
Limite_Inferior_iqr2021 = percentile25 - 1.5*iqr
print("Limite Superior permitido", Limite_Superior_iqr2021)
print("Limite Inferior permitido", Limite_Inferior_iqr2021)
```

```python
#Obtenemos datos limpios del DataFrame
data2021_iqr = cuantitativas2021[(y<=Limite_Superior_iqr2021)&(y>=Limite_Inferior_iqr2021)]
data2021_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos2021=data2021_iqr.isnull().sum()
valores_nulos2021
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data2021_1_iqr=data2021_iqr.copy()
data2021_1_iqr=data2021_1_iqr.fillna(round(data2021_iqr.mean(),1))
data2021_1_iqr
```

```python
#Unimos la columnas CUALITATIVAS con el dataframe

Datos_limpios2021 = pd.concat([cualitativas2021, data2021_1_iqr], axis=1)
Datos_limpios2021
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos2021=Datos_limpios2021.isnull().sum()
valores_nulos2021
```

```python
#Convertir DataFrame a CSV
Datos_limpios2021.to_csv("Valores Atipicos 2021 LIMPIOS.csv")
```
